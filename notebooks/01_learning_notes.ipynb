## 3. Score Prediction / 分数预测

我训练了第一个机器学习模型：根据学习时间预测考试分数。

这是一个回归任务，因为模型要预测的是一个具体数字。

重要概念：

- `X`: 特征，也就是模型输入。
- `y`: 标签，也就是模型要预测的目标。
- `fit`: 训练模型。
- `predict`: 使用模型预测。
- `train_test_split`: 划分训练集和测试集。
- `mean_absolute_error`: 平均绝对误差，用来评估模型预测差多少。

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

scores_df = pd.read_csv("../data/study_scores.csv")
scores_df
````

````python
X = scores_df[["study_hours"]]
y = scores_df["score"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)

test_predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, test_predictions)

round(mae, 2)
````

````python
new_data = pd.DataFrame({"study_hours": [9]})
predicted_score = model.predict(new_data)[0]
predicted_score = min(predicted_score, 100)

round(predicted_score, 2)
````
